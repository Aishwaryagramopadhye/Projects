#  Exploratory Data Analysis

# Attrition with respect to department

```python
sns.countplot(x='Department',hue='Attrition',data=df)
plt.title("Attrition w.r.t. department")
plt.show
```

# attrition with respect to EducationField 

```python
sns.countplot(x='EducationField',hue='Attrition',data=df)
plt.title("Attrition w.r.t. EducationField ")
plt.xticks(rotation=45)
plt.show
```

# Attrition with respect to job role

```python
plt.figure(figsize=(15,6))
sns.countplot(x='JobRole',hue='Attrition',data=df)
plt.xticks(rotation=45)
plt.title("Attrition w.r.t.job role")
plt.show
```

# Attrition with respect to gender

```python
sns.countplot(x='Gender',hue='Attrition',data=df,palette='hot')

plt.title('Attrition w.r..t. Gender')
```

```python
# Findings:
# Males are more under attrition than females
```
