# Group By

```python
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values (by='Qty',ascending=False)
```

```python
var=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values (by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=var)
```

```python
#From above graph you can see that most of th qty buys M size in the sales
```
